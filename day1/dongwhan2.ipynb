```python
import dash
import jupyter_dash
from dash import dcc
from dash import html
import plotly.express as px
from dash.dependencies import Input, Output

# Create the dash app
app = jupyter_dash.JupyterDash(__name__)

# Define the app layout
app.layout = html.Div(
    [
        html.H1("My Dashboard", style={"textAlign": "center"}),
        html.Div(
            [
                dcc.Graph(
                    id="graph1",
                    figure=px.scatter(x=[1, 2, 3], y=[4, 5, 6], title="Graph 1"),
                    style={
                        "height": "25vh", 
                        "width": "22%", 
                        "float": "left",
                        'border': '1px solid #ccc',
                        'border-radius': '5px',
                        'box-shadow': '2px 2px 5px #ccc',
                        'padding': '10px',
                        'margin': '10px',
                        'display': 'inline-block',
                    
                    },
                ),
                dcc.Graph(
                    id="graph2",
                    figure=px.line(x=[1, 2, 3], y=[4, 5, 6], title="Graph 2"),
                    style={
                        "height": "25vh", 
                        "width": "22%", 
                        "float": "left",
                        'border': '1px solid #ccc',
                        'border-radius': '5px',
                        'box-shadow': '2px 2px 5px #ccc',
                        'padding': '10px',
                        'margin': '10px',
                        'display': 'inline-block',
                    
                    },
                ),
                dcc.Graph(
                    id="graph3",
                    figure=px.bar(x=[1, 2, 3], y=[4, 5, 6], title="Graph 3"),
                    style={
                        "height": "25vh", 
                        "width": "22%", 
                        "float": "left",
                        'border': '1px solid #ccc',
                        'border-radius': '5px',
                        'box-shadow': '2px 2px 5px #ccc',
                        'padding': '10px',
                        'margin': '10px',
                        'display': 'inline-block',
                    
                    },
                ),
                dcc.Graph(
                    id="graph4",
                    figure=px.area(x=[1, 2, 3], y=[4, 5, 6], title="Graph 4"),
                    style={
                        "height": "25vh", 
                        "width": "22%", 
                        "float": "left",
                        'border': '1px solid #ccc',
                        'border-radius': '5px',
                        'box-shadow': '2px 2px 5px #ccc',
                        'padding': '10px',
                        'margin': '10px',
                        'display': 'inline-block',
                    
                    },
                ),
            ],
            style={"width": "100%", "display": "flex", "flex-wrap": "wrap"},
            className='graph-container'
        ),
        html.Div([
                 dcc.Graph(
                    id="graph4",
                    figure=px.area(x=[1, 2, 3], y=[4, 5, 6], title="Graph 4"),
                    style={
                        "height": "50vh", 
                        "width": "47%", 
                        "float": "left",
                        'border': '1px solid #ccc',
                        'border-radius': '5px',
                        'box-shadow': '2px 2px 5px #ccc',
                        'padding': '10px',
                        'margin': '10px',
                        'display': 'inline-block',
                    
                    },
                ),           
            
            html.Div([
                dcc.Dropdown(
                    id='dropdown',
                    options=[
                        {'label': 'Data 4', 'value': 'data4'},
                        {'label': 'Data 5', 'value': 'data5'},
                        {'label': 'Data 6', 'value': 'data6'}
                    ],
                    value='data4'
                ),
                dcc.Graph(id='graph4')
            ], 
                    style={
                    "height": "50vh", 
                    "width": "47%", 
                    "float": "left",
                    'border': '1px solid #ccc',
                    'border-radius': '5px',
                    'box-shadow': '2px 2px 5px #ccc',
                    'padding': '10px',
                    'margin': '10px',
                    'display': 'inline-block',
                    
                },
            
                className='six columns'),
        ],
        ),
    ],

    style={
        "width": "80%",
       
        "background-color": "#F8F8F8",
        "font-family": "Arial",
        "color": "#333",
        "padding": "20px",
    },
)

@app.callback(Output('graph4', 'figure'), [Input('dropdown', 'value')])
def update_graph4(selected_value):
    if selected_value == 'data4':
        data = [{'x': [1, 2, 3], 'y': [3, 1, 4], 'type': 'bar', 'name': 'Data 4'}]
    elif selected_value == 'data5':
        data = [{'x': [1, 2, 3], 'y': [5, 2, 3], 'type': 'scatter', 'name': 'Data 5'}]
    else:
        data = [{'x': [1, 2, 3], 'y': [2, 4, 1], 'type': 'bar', 'name': 'Data 6'}]

    return {
        'data': data,
        'layout': {
            'title': 'Graph 4'
        }
    }



if __name__ == "__main__":
    app.run_server(mode='jupyterlab', dev_tools_hot_reload=True)
```
